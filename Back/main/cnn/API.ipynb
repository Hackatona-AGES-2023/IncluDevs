```python
def base64_to_image(imagem64):
    import base64
    import io
    from matplotlib import pyplot as plt
    import matplotlib.image as mpimg

    i = base64.b64decode(imagem64)
    i = io.BytesIO(i)
    i = mpimg.imread(i, format='JPG')
    
    return i
```
